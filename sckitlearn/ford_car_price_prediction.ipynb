```python
X = df.drop('price', axis=1) # to separate the features from the target variable
y = df['price']
```

```python
df.columns
```

```python
X_one_encoded = pd.get_dummies(X, columns=['model','fuelType', 'transmission'], drop_first=True) # to encode the categorical features using one-hot encoding
```

```python
X_one_encoded = X_one_encoded.astype(int)
```

```python
X_one_encoded
```

```python
from sklearn.preprocessing import LabelEncoder


columns=['model','fuelType', 'transmission']

Xlable = X.copy()
label_encoder = {}

for col in columns:
    le = LabelEncoder()
    Xlable[col] = le.fit_transform(Xlable[col].astype(str))
    label_encoder[col] = le
```

```python
Xlable
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
numericals_columns=['year', 'mileage', 'engineSize', 'tax', 'mpg']
scaler = StandardScaler()
X_one_encoded[numericals_columns] = scaler.fit_transform(X_one_encoded[numericals_columns])
```

```python
X_one_encoded
```

```python
print(Xlable.columns)
```

```python
Xlable[[int['model', 'year', 'transmission', 'mileage', 'fuelType', 'tax', 'mpg',
       'engineSize']]] = Xlable[[int['model', 'year', 'transmission', 'mileage', 'fuelType', 'tax', 'mpg',
       'engineSize']]]
```

```python
Xlable
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_one_encoded, y, test_size=0.2, random_state=42) # to split the dataset into training and testing sets
```

```python
model = LinearRegression() # to create an instance of the Linear Regression model
model.fit(X_train, y_train) # to train the model on the training data
```

```python
y_pred = model.predict(X_test) # to make predictions on the test set
```

```python
y_pred
```

```python
y_test
```

```python
r2 = r2_score(y_test, y_pred) # to calculate the R-squared score
print(f'R-squared score: {r2:.4f}')
```

```python
n = X_test.shape[0]
p = X_test.shape[1]
adjusted_r2 = 1 - (1 - r2) * (n - 1) / (n - p - 1) # to calculate the adjusted R-squared
print(f'Adjusted R-squared score: {adjusted_r2:.4f}')
```
