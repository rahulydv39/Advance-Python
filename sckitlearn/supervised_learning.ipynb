```python
df
```

```python
inertia = []
k_range = range(1, 11)
```

```python
inertia
```

```python
kmeans_final = KMeans(n_clusters=3, random_state=42)
```

```python
cluster_labels = kmeans_final.fit_predict(X_scaled)
```

```python
cluster_labels
```

```python
df['cluster'] = cluster_labels
```

```python
sns.scatterplot(data=df, x='Feature 1', y='Feature 2', hue='cluster', palette='viridis')
```

### 2. DB SCAN

```python
from sklearn.datasets import make_moons
```

```python
X, y_true = make_moons(n_samples=300, noise=0.05, random_state=42)
```

```python
from sklearn.cluster import DBSCAN, KMeans
```

```python
df = pd.DataFrame(X, columns=['Feature 1', 'Feature 2'])
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df)
```

```python
kmeans = KMeans(n_clusters=2, random_state=42)
kmeans_labels = kmeans.fit_predict(X_scaled)
```

```python
df['kmeans_cluster'] = kmeans_labels
```

```python
sns.scatterplot(data=df, x='Feature 1', y='Feature 2', hue='kmeans_cluster', palette='Set1')
```

```python
dbscan = DBSCAN(eps=0.3, min_samples=5)
dbscan_labels = dbscan.fit_predict(X_scaled)
```

```python
df['label'] = dbscan_labels
```

```python
sns.scatterplot(data=df, x='Feature 1', y='Feature 2', hue='label', palette='Set1')
plt.title('DBSCAN Clustering Results')
plt.show()
```

## Dimensity Reduction through PCA

```python
import pandas
from sklearn.decomposition import PCA  
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import make_blobs
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
X , y = make_blobs(n_samples=500, centers=5, cluster_std=1.5, random_state=42)
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)
```

```python
X_pca
```

```python
df_pca = pd.DataFrame(X_pca, columns=['Principal Component 1', 'Principal Component 2'])
df_pca['Cluster'] = y
```

```python
sns.scatterplot(data=df_pca, x='Principal Component 1', y='Principal Component 2', hue='Cluster', palette='Set2')
plt.title('K-Means Clustering Results')
plt.show()
```
